```python
import os
from huggingface_hub import login
from detoxify import Detoxify
from dotenv import load_dotenv, find_dotenv
```

```python
def get_key(key_name):
    _ = load_dotenv(find_dotenv())
    key = os.getenv(key_name)
    return key
```

```python
login(get_key('HUGGING_FACE_TOKEN'))
```

```python
def toxicity_predict(text):
    toxicity_scores = Detoxify('original').predict(text)
    print(toxicity_scores)
```

```python
toxicity_predict("You're a worthless piece of trash.")
```

```python
toxicity_predict("Do you like chicken nuggets?")
```
